```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/train.csv")
```

```python
data
```

```python
unsurvive_data = data[data["Survived"] == 0]
survive_data = data[data["Survived"] == 1]
```

```python
un_p = unsurvive_data["Pclass"].value_counts()
p      =  survive_data["Pclass"].value_counts()
result = pd.concat([un_p.rename("Pclass unsurvive"),p.rename("Pclass survive")],axis=1)
result.plot.bar()
```
